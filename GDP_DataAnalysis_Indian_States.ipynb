```python
#Import the libraries for reading directories & char encoding detection
import os
import chardet
```

```python
#Define a function to check and return the file encoding
def get_file_encoding(file_name):
    csv_content = open(file_name,'rb').read()
    char_enc=chardet.detect(csv_content)['encoding']
    return char_enc
```

```python
#Define an empty dataframe, which would be used to hold all state data from CSV files
allstate_df=pd.DataFrame()

#Set the path variable for reading the csv files
path = "./Part1B"         

#Define a for loop to check the directory and import csv files into dataframe
for entry in os.scandir(path):
    #Call the function get_fileencoding and assignt he file encoding to a variable
    csv_encoding=get_file_encoding(entry.path)
    
    #Read the csv into itr_df DataFrame
    itr_df=pd.read_csv(entry.path,encoding=csv_encoding)

    #Assign state name based by extracting it from file name
    itr_df['StateName']=entry.path.split('-')[1]
    
    #Concat the datafram into allstate_df
    allstate_df=pd.concat([allstate_df,itr_df],sort=True)
```

```python
#Extract the 3 sectors' data & GSDP data into a new DataFra,e
get_sect_df=allstate_df.loc[allstate_df['Item'].isin(['Primary','Secondary','Tertiary','Gross State Domestic Product'])][['StateName','Item','2014-15']]
```

```python
get_sect_df.head()
```

```python
# Create the sector dataframe by pivoting get_sect_df DataFrame
sector_df=get_sect_df.pivot(index='StateName', columns='Item', values='2014-15')
sector_df.head()
```

```python
#Sort the data in sector_df dataframe based on GSDP
sector_df.sort_values(by='Gross State Domestic Product',ascending = False,inplace=True)
```

```python
# Create a column and assign the Primary sector % contribution 
sector_df['Primary%']=round((sector_df['Primary']/sector_df['Gross State Domestic Product'])*100,2)

# Create a column and assign the Secondary sector % contribution 
sector_df['Secondary%']=round((sector_df['Secondary']/sector_df['Gross State Domestic Product'])*100,2)

# Create a column and assign the Tertiary sector % contribution 
sector_df['Tertiary%']=round((sector_df['Tertiary']/sector_df['Gross State Domestic Product'])*100,2)

# Create a column and assign the Tax+Subsidy sector % contribution 
sector_df['Tax&Subsidy%']=round((100-sector_df['Primary%']-sector_df['Secondary%']-sector_df['Tertiary%']),2)
```

```python
# Drop the columns 'Gross State Domestic Product','Primary','Secondary','Tertiary'
sector_df.drop(['Gross State Domestic Product','Primary','Secondary','Tertiary'],axis=1,inplace=True)
```

```python
sector_df.head()
```

```python
#Plot visualisation for the sector wise contribution to State GSDP
sector_df.plot(kind='bar', stacked=True, figsize=(16,8))

#Give Label to the y-axis
plt.ylabel('Percentage of GSDP')

#Provide title to the plot
plt.title('Sector-wise % contribution to GSDP during 2014-15 for all States'+"\n"+'Ordered using State with Highest GSDP to Lowest')

#Provide legend location
plt.legend(loc='upper right',title='Sector')
```

### Create and visualise a scatter plot for each grade of school drop out % against the Per_Capita_GSDP_2014-15


```python
#Define the dictionary for keywords to be used in plotting
kws = dict(s=100, linewidth=0.5, edgecolor="k")

#Create a seaborn FacetGrid with Dataframe and plot State Category as hue
g = sns.FacetGrid(drops_and_gsdp_df, hue="State_Ctgy", palette="Set1", height=5, aspect=2)

#Label the plot
plt.title("Per Capita GSDP vs. Primary Drop out %")

#Generate a scatter plot with Per Capita & Primary droput %
g = (g.map(plt.scatter, "Per_Capita_GSDP_2014-15", "Primary - 2014-2015", **kws).add_legend())
```

-  <font color=blue> From the Per Capita vs Primary Drop out % plot we see above, it is evident that for the states in C1 & C2 (which have higher Per Capita than C3 & C4) category, the drop out percentage is low at Primary class compared to the states in C3 & C4 categories. Therefore, with increase in Per Capita GSDP, the drop out percentage can be said to be coming down. </font>

<font color=blue>From the Per Capita vs Senior Secondary Drop out % plot we see above, ironically the C1 category states have higher drop out percentage than C2, C3 & C4 states.This can be attributed the fact that most of the students drop out at early stages in C3 & C4 or in some caes C2 states also. But when it comes to C1 states, they continue up until Senior Secondary before dropping out.</font>

```python
#Create a correlation Matrix for the Per Capita GSDP against the school dropt out rate 
dropouts_corr=drops_and_gsdp_df[['Per_Capita_GSDP_2014-15','Primary - 2014-2015','Upper Primary - 2014-2015','Secondary - 2014-2015','Senior Secondary - 2014-2015']].corr()

#Print the Correlation matrix
dropouts_corr
```

```python
#Generate a heapmap to visualise the correlation betwwen the values

cmap = sns.cm.rocket_r
fig, ax = plt.subplots(figsize=(10,8))
sns.heatmap(dropouts_corr,annot=True,cmap = cmap)
```

### From the Heatmap generated using the correlation matrix, we have the following observations:

-  <font color=blue> The Per Capita GSDP has negative correlation with Primary, Upper Primary & Secondary drop out percentage i.e. less number of students are dropping out at these 3 levels as the Per Capita GSDP increases. So for states with Low Per Capita GSDP, higher chances of students dropping out at Primary level.</font>

-  <font color=blue> The Per Capita GSDP correlation with Primary drop out percentage is -0.46 i.e. with every 1% increase in Per Capita GSDP, the Primary drop out rate can decrease by 0.46 %. For Upper Primary drop out, the number stands at decrease by 0.54% and Secondary drop outs decrease by 0.47 %.</font>

-  <font color=blue> The Per Capita GSDP has positive correlation with Senior Secondary drop out percentage i.e. more number of students are dropping out at this levels as the Per Capita GSDP increases. Around 0.33% increase in drop out % at Senior Secondary for every 1% increase in the Per Capita GSDP.</font>
    
-  <font color=blue> The way this should be looked at is - "For states with higher Per Capita GSDP, there is a higher chance of students continuing education up until Senior Secondary, but at the same time they are dropping out at that level"</font>
